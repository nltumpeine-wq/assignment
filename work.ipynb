# BUSINESS UNDERSTANDING
Context Cardiotocography (CTG) is widely used during pregnancy to monitor fetal heart rate and uterine contractions, serving as a critical tool to prevent fetal mortality and morbidity. This dataset provides numerical features extracted from CTG exams to help build predictive machine learning classification models for fetal health.

Content The dataset contains 2,126 records of fetal cardiotocograms. The automated system extracted 35 numerical features (labeled V1 through V35), which include metrics like FHR baseline (beats per minute) and the number of accelerations per second. The final column, Class, is the target variable representing the consensus classification label assigned by expert obstetricians (ranging from 1 to 10 representing the FHR pattern class code).
